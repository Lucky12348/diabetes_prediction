```python
df_copy = df.drop(columns=["PatientID"], errors='ignore')
```

```python


# pour chaque colonne numérique
for col in df_copy.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot de {col}")
    plt.xlabel(col)
    plt.show()
    
    # valeurs extremes sur seruminsulin
    # same for tricepsthinkness (mais moindre)
```

```python
for col in df_copy.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], bins=30, kde=True)  
    plt.title(f"Histogramme de {col}")
    plt.xlabel(col)
    plt.ylabel("Fréquence")
    plt.show()
```
